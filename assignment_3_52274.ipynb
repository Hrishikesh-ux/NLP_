```python
import spacy
from collections import Counter
import subprocess
import sys

# Attempt to load the model, if not found, download it
try:
    nlp = spacy.load("en_core_web_sm") # Changed to specific model name
except OSError:
    print("Downloading en_core_web_sm model...")
    # Use subprocess to run the pip install command for the spacy model package
    subprocess.check_call([sys.executable, "-m", "pip", "install", "en_core_web_sm"])
    nlp = spacy.load("en_core_web_sm") # Load the specific model name after installation

text = "Loving the new AI features ðŸ˜‚ #AI #MachineLearning"
doc = nlp(text)
pos_counts = Counter([token.pos_ for token in doc])

nouns = []
verbs = []

for token in doc:
  if token.pos_ in ["NOUN", "PROPN"]:
    nouns.append(token.text)
  elif token.pos_ == "VERB":
    verbs.append(token.text)

noun_freq = Counter(nouns)
verb_freq = Counter(verbs)

print("Noun Frequencies:", noun_freq)
print("Verb Frequencies:", verb_freq)
```

spaCy Part-of-Speech Tagging and Frequency Analysis for assignment3_text (from JSON file)

```python
import spacy
from collections import Counter
import subprocess
import sys

# Attempt to load the model, if not found, download it
try:
    nlp = spacy.load("en_core_web_sm")
except OSError:
    print("Downloading en_core_web_sm model...")
    subprocess.check_call([sys.executable, "-m", "pip", "install", "en_core_web_sm"])
    nlp = spacy.load("en_core_web_sm")

# Using the text extracted from the JSON for Assignment 3
text_for_analysis = assignment3_text

doc = nlp(text_for_analysis)

print("SpaCy POS Tags (Token, Universal POS, Detailed Tag):")
for token in doc:
    print(token.text, token.pos_, token.tag_)

pos_counts = Counter([token.pos_ for token in doc])

nouns = []
verbs = []

for token in doc:
  if token.pos_ in ["NOUN", "PROPN"]:
    nouns.append(token.text)
  elif token.pos_ == "VERB":
    verbs.append(token.text)

noun_freq = Counter(nouns)
verb_freq = Counter(verbs)

print("\nNoun Frequencies:", noun_freq)
print("Verb Frequencies:", verb_freq)
```
